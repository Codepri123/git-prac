```python
def find_missing(nums):
    expected_output=(max(nums))*(max(nums)+1)//2
    missing=expected_output-sum(nums)
    return missing
print(find_missing([3,0,1]))
```

```python
def compress_string(text):
 freq={}
 for i in text:
    if i in freq:
      freq[i]+=1
    else:
      freq[i]=1
 return freq
print(compress_string("helloo"))
```
